```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'b')
axes.set_xlabel('X coordinate')
axes.set_ylabel('Y coordnate')
axes.set_title('XY Coordinate')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2  = fig.add_axes([0.15,0.4,0.4,0.4])


axes1.plot(x,y,'b')
axes2.plot(y,x,'g')
```

```python
fig.savefig("graph.png")
```

```python
fig,axes = plt.subplots()
axes.plot(x,x**2,label='x square')
axes.plot(x,x**3,label='y square')

axes.legend(loc=0)
```
